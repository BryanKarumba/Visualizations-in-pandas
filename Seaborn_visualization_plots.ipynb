**Displot**
<br>Distribution plot is a graphical representation used to visualize the distribution of a dataset.It helps to understand the underlying pattern of data, including its central tendency, variability, skewness, and the presence of outliers.
<br> They include:
1. Histogram
2. KDE
3. Box plot
4. Violin Plot
5. Rug plot
6. Joint plot

```python
#Plotting a displot to show the distribution of total_bill from our dataset.
#kde(Kernel density Function)- you can call it as true to see the kde line
sns.displot(tips['tip'],kde=True)
```

*JointPlot*
<br>A joint plot is a type of data visualization that simultaneously displays the distribution of two variables and their potential relationship

```python
#Here we will see the distribution of total bill and tip hence we can see their relationship.
sns.jointplot(x='total_bill',y='tip', data=tips)
```

```python
#Adding the display as a hexagons
sns.jointplot(x='total_bill',y='tip', data=tips,kind='hex')
```

```python
#adding a regression line
sns.jointplot(x='total_bill',y='tip', data=tips,kind='reg')
```

```python
#adding the kde as the kind of plot we want.
sns.jointplot(x='total_bill',y='tip', data=tips,kind='kde')
```

**VIOLIN PLOTS**
<br> This allows you to easily compare the distributions between groups.
<br>**Shape of the Violin:** The width of the violin plot represents the density of the data at different score levels. Wider sections indicate more data points, while narrower sections indicate fewer data points.

<br>**Center Line:** The thick black line inside each violin plot represents the interquartile range (IQR) of the data distribution. It spans from the 25th percentile (lower quartile) to the 75th percentile (upper quartile). This line gives you an idea of the central tendency of the data.

<br>**White Dot or Line:** This indicates the median of the data distribution, which is the middle value when all scores are arranged in ascending order.

<br>**Extended Lines (Whiskers):** The width of the violin plot gradually tapers off towards the minimum and maximum values of the data. This gives you a sense of the spread of the data.

<br>**Color:** If you're using a violin plot with multiple groups (e.g., different schools), each group's violin will have a different color. This allows you to easily compare the distributions between groups.

```python
#shows the kernel density distribution-
sns.violinplot(x='day',y='total_bill',data=tips)
```

**STRIPPLOT**
<br>Strip plots are a type of data visualization technique used to display the distribution of a continuous variable. In a strip plot, each data point is represented as a point along a single axis, usually the x-axis, with all points aligned vertically.
<br>They're useful when you have a relatively small dataset and want to understand how the data is distributed

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

**MATRIX PLOTS**
PRIMARILY HEATMAPS

```python
#Here we import the flights dataset inbuilt in jupyter notebook and the flights dataset
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
flight=sns.load_dataset('flights')
flight.head()
```

The data should be in matrix form to ensure the heatmap to be plotted!
<br>We can use the pivoting method

```python
#Pivot method. Herein i will convert the flight data to a matrix form for us to be able to draw a heatmmap from it.
fp=flight.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
#generating the heatmap.
sns.heatmap(fp)
```

```python
#We have different styles for cmap including coolwarm, YlGnBu, Blues, etc., and their reversed versions would be coolwarm_r, YlGnBu_r, Blues_r,
# In order to reverse the color to show light colors for lower numbers, add a suffix ('_r') to the type of cmap you call
sns.heatmap(fp,cmap='magma_r')
```

```python
#Customizing the heatmap further
sns.heatmap(fp,cmap='magma_r',linewidth=1,linecolor='yellow')
```

```python
sns.heatmap(fp,cmap='coolwarm',linewidth=0)
```

**CLUSTERMAPS**
<BR>This is used when you want to understand the structure and relationships within the data. 
<br>**Identifying Patterns:** By clustering rows and columns, a clustermap makes it easier to see groups of similar data points. For instance, in gene expression data, genes with similar expression patterns across conditions can be grouped together.<br>
**Detecting Anomalies:** Outliers or unusual patterns stand out more clearly when similar data points are clustered together.<br>
**Data Reduction:** By focusing on clusters rather than individual data points, you can simplify complex data and identify broader trends or categories

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
#We can actually add a normalized scale in the data
#this changes the scale in the top left corner form 0 to 1
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

**REGRESSION PLOTS**
<br>They help you to understand the relationship between variables by plotting data points and fitting a regression line
<br> There are different types of regression plots:Simple linear regression plots, Residual plots, Polynomial plots and Multiple regression plots

```python
# A linear model plot
sns.set_context('notebook')
sns.lmplot(x='tip',y='total_bill',data=tips)
```

```python
#Here we add a hue i.e more filtering.
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex')
```

```python
#addind separation using columns(col), that is by either sex or day
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```
