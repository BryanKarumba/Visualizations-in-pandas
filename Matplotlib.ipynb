# MATPLOTLIB FOR VISUALIZATION
The measurements here are usually in inches

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
#functional:
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Measurement')
```

```python
#object oriented methods- here we have control over the axes lengths 
#x=np.linspace(start,stop)
x=np.linspace(0,4)
y=x**2
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
#Creating a plot inside of another plot
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # the 0.5 is 50% of the figurej
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.3,0.4,0.3])
```

```python
#Manually setting the dimensions of the plots.
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Big Plot')
axes2.set_title('Smaller Plot')
```

```python
#You can combine the two:Creating multiple 
fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(6,2))

plt.tight_layout() #This esures plots are well spaced
```

# Figure Size Ratio and DPI(Dots per inch)

```python
fig=plt.figure(figsize=(8,2))
#8 is the width and 2 is the height

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Plot of Squares of Numbers')
ax.set_ylabel('Y-axis')
ax.set_xlabel('X-axis')
```

# Legends

```python
#Plotting the axes on one plot.
fig1=plt.figure(figsize=(4,2))

ax=fig1.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
#Adding legends to the plot.
fig1=plt.figure(figsize=(4,2))

ax=fig1.add_axes([0,0,1,1])
ax.plot(x,y, label='x vs y')
ax.plot(y,x, label='y vs x')

ax.legend(loc=0)# adds the label to the best location in the plot, hence avoiding it from overlapping

#here, we have added labels to the axes of the coloured lines.(i.e legend)
```

# Customizing the Plots
1. Color
2. Linewidh
3. Alpha(transparency)
4. Linestyle
5. Markers

```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])

ax.plot(x,y, color='green') #adding color name. 
ax.plot(x,y, linewidth=4) #adding linewidth
ax.plot(x,y, alpha=1)
```

LINESTYLE


```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,2,1])


ax.plot(x,x**2, linewidth=4,linestyle='--') #adding linestyle
```

```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,4,2])
ax.set_title('Plot customization')

ax.plot(x,y, color='green') #adding color name. 
ax.plot(x,y, linewidth=1, marker='o', markersize=10, markerfacecolor='yellow', markeredgecolor='green')
```
